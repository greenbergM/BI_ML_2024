### Задание 2. Обучение дерева решений (10 баллов)

Предобработайте признаки так, чтобы их можно было подать в модель дерева решений и обучите ее, постарайтесь подобрать оптимальные гиперпараметры, которые вам известны. Постарайтесь также обосновать выбор метрики в данном случае. При подборе гиперпараметров может помочь `GridSearchCV` [ссылка](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html).

**С учетом несбалансированности классов наиболее предпочтительной является метрика f1-score. При этом я буду использовать взвешенную версию (average=weighted), которая учитывает пропорции классов (разница между самым большим и маленьким в нашем случае >10 раз)**

**Сделаю пайплайн**

```python
cat_cols = ['key']
num_cols = ['Popularity', 'danceability', 'energy', 'loudness', 'mode',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'duration_in min/ms'] 
# mode уже бинарная, так что я оставлю эту переменную в численных - ее прогонять через onehotencoder не буду

all_features = cat_cols + num_cols
X = data[all_features]
y = data['Class'].values
```

```python
preprocessor = ColumnTransformer(transformers=[
    ("scaler", StandardScaler(), num_cols),
    ("ohe", OneHotEncoder(drop="first"), cat_cols)
])


treeclass_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("tree_classifier", DecisionTreeClassifier())])

treeclass_pipeline
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
param_grid = {
    "tree_classifier__max_depth": [3, 5, 7, 10, 13, 15],
    "tree_classifier__min_samples_split": [2, 5, 10, 15],
    "tree_classifier__min_samples_leaf": [1, 2, 4, 10, 15],
}

grid_search = GridSearchCV(treeclass_pipeline, param_grid, cv=5, scoring='f1_weighted')

grid_search.fit(X_train, y_train)

print("Best params:", grid_search.best_params_)
print("Best f1-score:", grid_search.best_score_)
```

```python
tree_model = DecisionTreeClassifier(max_depth=10, min_samples_leaf=2, min_samples_split=15)
tree_model.fit(X_train, y_train)
y_pred = tree_model.predict(X_test)

print(f'f1-score for test:{f1_score(y_test, y_pred, average='weighted')}') 
```

**Получилось так себе...**

```python
tree.plot_tree(tree_model) #Каждый начинающий биоинформатик должен построить дерево, вырастить дом и посадить сына
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(conf_matrix).plot()
plt.grid(False);
```

**По всей видимости проблема действительно в дисбалансе классов, т.к. модель охотнее предсказывает другие классы как 6,9 и в особенности 10 классы - как раз 3 самых часто встречающихся класса.**
